```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5]
y = [1,2,3,4,5]
```

```python
plt.xlabel('x')
plt.ylabel('y')
plt.title('Graphs title is x and y')
plt.plot(x,y)
plt.show()
```
